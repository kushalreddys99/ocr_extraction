```python

%%writefile app.py

from fastapi import FastAPI, UploadFile, File, Form
from fastapi.middleware.cors import CORSMiddleware
import tempfile
import json
import ollama
import os
import time

app = FastAPI()

app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=False,
    allow_methods=["*"],
    allow_headers=["*"],
)


@app.get("/")
def home():
    return {
        "status": "running"
    }


@app.get("/cors-test")
def cors_test():
    return {
        "message": "cors works"
    }


@app.post("/extract")
async def extract(
    file: UploadFile = File(...),
    prompt: str = Form(...)
):
    try:
        print("=" * 60)
        print("Received File:", file.filename)
        print("Received Prompt:")
        print(prompt)
        print("=" * 60)

        with tempfile.NamedTemporaryFile(
            delete=False,
            suffix=".jpg"
        ) as temp:
            temp.write(await file.read())
            img_path = temp.name

        start_time = time.time()

        response = ollama.chat(
            model="qwen2.5vl:7b",
            messages=[
                {
                    "role": "user",
                    "content": prompt,
                    "images": [img_path]
                }
            ],
            format="json"
        )

        raw_output = response["message"]["content"]

        processing_time = round(
            time.time() - start_time,
            2
        )

        print(
            f"Model Processing Time: {processing_time} sec"
        )

        print("Raw Model Output:")
        print(raw_output)

        try:
            result = json.loads(raw_output)

        except Exception as parse_error:
            result = {
                "success": False,
                "error": "Invalid JSON returned by model",
                "raw_output": raw_output,
                "parse_error": str(parse_error)
            }

        result["model_processing_time"] = processing_time

        try:
            os.remove(img_path)
        except:
            pass

        return result

    except Exception as e:
        print("EXTRACTION ERROR:")
        print(str(e))

        return {
            "success": False,
            "error": str(e)
        }

```

```python
import ollama
import json
import pandas as pd
import time
import os
import re
from PIL import Image

# ============================================================
# IMAGE PATHS
# ============================================================

image_paths = [
    "/kaggle/input/datasets/kushalreddys/abcdef/55877.jpg",
    "/kaggle/input/datasets/kushalreddys/abcdef/55878.jpg",
    "/kaggle/input/datasets/kushalreddys/abcdef/55879.jpg"
]

# ============================================================
# PROMPT
# ============================================================

extraction_prompt = """
You are an OMR reader.

TASK:
Analyze ONLY Part-4 (भाग-4) of the OMR sheet image.

STRICT RULES:

1. Ignore Part-1, Part-2 and Part-3 completely.
2. Ignore all handwritten text.
3. Ignore signatures.
4. Ignore printed digits written inside boxes.
5. Extract values ONLY from filled OMR bubbles.
6. Never guess missing digits.
7. Never infer values from visible text.

ROLL NUMBER:

Locate:
"परीक्षार्थी का अनुक्रमांक"

For each column:

- Row 0 = digit 0
- Row 1 = digit 1
- Row 2 = digit 2
- Row 3 = digit 3
- Row 4 = digit 4
- Row 5 = digit 5
- Row 6 = digit 6
- Row 7 = digit 7
- Row 8 = digit 8
- Row 9 = digit 9

Choose ONLY the darkest filled bubble.

Concatenate digits from left to right.

BARCODE NUMBER:

Read ONLY the printed vertical barcode number beside the barcode.

OUTPUT:

{
  "barcode_number": "",
  "roll_number": ""
}

Return ONLY JSON.
No markdown.
No explanation.
"""

# ============================================================
# FUNCTIONS
# ============================================================

def crop_part4(image_path):
    """
    Crop only the right-most section (Part-4)
    """

    img = Image.open(image_path)

    width, height = img.size

    # Right-most third
    x1 = int(width * 0.66)

    part4 = img.crop(
        (
            x1,
            0,
            width,
            height
        )
    )

    # enlarge for better OMR visibility
    part4 = part4.resize(
        (
            part4.width * 2,
            part4.height * 2
        ),
        Image.LANCZOS
    )

    temp_path = "/kaggle/working/temp_part4.jpg"
    part4.save(temp_path, quality=100)

    return temp_path


def extract_json(text):

    text = text.strip()

    try:
        return json.loads(text)
    except:
        pass

    match = re.search(r'\{.*\}', text, re.DOTALL)

    if match:
        return json.loads(match.group())

    raise ValueError("JSON not found")


# ============================================================
# PROCESSING
# ============================================================

all_records = []

for image_path in image_paths:

    print(f"\nProcessing: {image_path}")

    start = time.time()

    try:

        cropped_image = crop_part4(image_path)

        response = ollama.chat(
            model="qwen2.5vl:7b",
            messages=[
                {
                    "role": "user",
                    "content": extraction_prompt,
                    "images": [cropped_image]
                }
            ],
            format="json",
            options={
                "temperature": 0,
                "num_ctx": 8192
            }
        )

        content = response["message"]["content"]

        try:
            data = extract_json(content)

        except:

            print("Retrying once...")

            response = ollama.chat(
                model="qwen2.5vl:7b",
                messages=[
                    {
                        "role": "user",
                        "content": extraction_prompt,
                        "images": [cropped_image]
                    }
                ],
                format="json",
                options={
                    "temperature": 0,
                    "num_ctx": 8192
                }
            )

            content = response["message"]["content"]
            data = extract_json(content)

        data["file_source"] = os.path.basename(image_path)

        all_records.append(data)

        elapsed = time.time() - start

        print(
            f"✓ Done in {elapsed:.2f}s"
        )

        print(data)

    except Exception as e:

        elapsed = time.time() - start

        print(
            f"✗ Failed after {elapsed:.2f}s"
        )

        print(e)

# ============================================================
# SAVE EXCEL
# ============================================================

if len(all_records):

    df = pd.DataFrame(all_records)

    preferred_columns = [
        "file_source",
        "barcode_number",
        "roll_number"
    ]

    existing = [
        c for c in preferred_columns
        if c in df.columns
    ]

    remaining = [
        c for c in df.columns
        if c not in existing
    ]

    df = df[existing + remaining]

    output_excel = (
        "/kaggle/working/Strict_Part4_OMR_Data.xlsx"
    )

    df.to_excel(
        output_excel,
        index=False,
        engine="openpyxl"
    )

    print("\nExcel saved:")
    print(output_excel)

else:

    print("\nNo records extracted.")

# ============================================================
# CLEANUP
# ============================================================

temp_file = "/kaggle/working/temp_part4.jpg"

if os.path.exists(temp_file):
    os.remove(temp_file)
```
